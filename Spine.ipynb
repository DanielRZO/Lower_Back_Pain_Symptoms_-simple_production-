# Machine Learning

## Separar bases de treino e teste

```python
# semente para reprodução de resultados
seed=10000

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(spine.loc[:,spine.columns != 'Class_att'], spine['Class_att'], test_size=0.2, random_state=seed)

print('shape input de treino: ', X_train.shape)
print('shape input de teste: ', X_test.shape)
print('shape output de treino: ', y_train.shape)
print('shape output de teste: ', y_test.shape)
```

## Árvore de Decisão

```python
# treinar modelo
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(min_samples_leaf=5, random_state=seed)
model.fit(X_train, y_train);
```

```python
# Avaliar modelo treinado
model.score(X_test, y_test) # acurácia
```

```python
y_pred = model.predict(X_test) #inferência do teste
```

```python
import numpy as np
unique_elements, counts_elements = np.unique(y_test, return_counts=True)
print("Frequência:")
print(np.asarray((unique_elements, counts_elements)))
```

```python
# Matriz de confusão
from sklearn.metrics import confusion_matrix
confMatrix = pd.DataFrame(confusion_matrix(y_pred, y_test), 
                          index=['Previsto Abnormal', 'Previsto Normal'], 
                          columns=['Real Abnormal', 'Real Normal'])
confMatrix
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(500, max_features=12, min_samples_leaf=4, random_state=seed);
rf.fit(X_train, y_train) 
```

```python
# Métrica acurácia
rf.score(X_test, y_test)
```

```python
y_pred_rf = rf.predict(X_test) #inferência do teste
```

```python
# Matriz de confusão
from sklearn.metrics import confusion_matrix
confMatrix = pd.DataFrame(confusion_matrix(y_pred_rf, y_test), 
                          index=['Previsto Abnormal', 'Previsto Normal'], 
                          columns=['Real Abnormal', 'Real Normal'])
confMatrix
```
